```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("../murders-feature-selected.csv")
display(data)
```

```python
data["Relationship"] = data["Relationship"].astype("category")
```

```python
X = data.drop("Relationship", axis=1)
y = data["Relationship"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
c45_model = DecisionTreeClassifier()
c45_model.fit(X_train, y_train)
c45_pred = c45_model.predict(X_test)
```

```python
c45_accuracy = accuracy_score(y_test, c45_pred)
print("C4.5 Accuracy:", c45_accuracy)
```
