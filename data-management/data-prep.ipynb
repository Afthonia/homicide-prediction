```python
# > 9m for all
data = pd.read_csv("../dataset-versions/homicides.csv")
display(data.head())
```

```python
dataDF = pd.DataFrame(data)
print(dataDF.info())
```

```python
allDuplicates = dataDF[dataDF.duplicated(keep=False)]
display(allDuplicates)
```

```python
duplicates = dataDF[dataDF.duplicated(keep='first')]
display(duplicates)
```

```python
dataDF.drop_duplicates(keep='first', inplace=True)
display(dataDF)
```

```python
df = dataDF.drop(['ID', 'CNTYFIPS', 'Ori', 'Agency', 'Incident', 'Subcircum', 'OffCount', 'FileDate', 'MSA'], axis='columns')
display(df)
```

```python
#print(df[df['VicAge'] == 999])
# 12672 records
df.loc[df['VicAge'] == 999, 'VicAge'] = np.nan

vic_mean = round(np.nanmean(df['VicAge'], axis=0))
print(vic_mean)
df['VicAge'].fillna(vic_mean, inplace=True)

df['VicAge'] = df['VicAge'].astype(int)
```

```python
#print(df[df['OffAge'] == 999])
# 289160 records
df.loc[df['OffAge'] == 999, 'OffAge'] = np.nan

off_mean = round(np.nanmean(df['OffAge'], axis=0))
print(off_mean)
df['OffAge'].fillna(off_mean, inplace=True)

df['OffAge'] = df['OffAge'].astype(int)
```

```python
print(df.info())
df.to_csv('../dataset-versions/homicides-new.csv', index=False)
```

```python
file.write("YEAR (Year - YearInterval)\n\n")

file.write('1976-1995 : 0\n')
file.write('1996-2010 : 1\n')
file.write('2011-2022 : 2\n')

df.loc[(df['Year'] >= 1976) & (df['Year'] <= 1995), 'Year'] = 0
df.loc[(df['Year'] >= 1996) & (df['Year'] <= 2010), 'Year'] = 1
df.loc[(df['Year'] >= 2011) & (df['Year'] <= 2022), 'Year'] = 2

df.rename(columns={'Year': 'YearInterval'}, inplace=True)

file.write('\n'*5)
```

```python
display(df)
```

```python
file.write('MONTH (Month -> Season)\n\n')

file.write("""
           December,
           January,
           February : 0\n
           """)
file.write("""
           March,
           April,
           May : 1\n
           """)
file.write("""
           June,
           July,
           August : 2\n
           """)
file.write("""
           September,
           October,
           November : 3\n
           """)


# Winter = 0
df.loc[(df['Month'] == 'December') | (df['Month'] == 'January') | (df['Month'] == 'February'), 'Month'] = 0

# Spring = 1
df.loc[(df['Month'] == 'March') | (df['Month'] == 'April') | (df['Month'] == 'May'), 'Month'] = 1

# Autumn = 2
df.loc[(df['Month'] == 'June') | (df['Month'] == 'July') | (df['Month'] == 'August'), 'Month'] = 2

# Summer = 3
df.loc[(df['Month'] == 'September') | (df['Month'] == 'October') | (df['Month'] == 'November'), 'Month'] = 3

df.rename(columns={'Month': 'Season'}, inplace=True)

file.write('\n'*5)
```
