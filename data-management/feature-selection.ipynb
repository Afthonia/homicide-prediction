```python
import pandas as pd
```

```python
data = pd.read_csv("../dataset-versions/murders-cleaned.csv")
display(data)
```

```python
label = 'Relationship'
```

```python
corr_matrix = data.corr()
corr_with_target = corr_matrix[label]
```

```python
k = 16
top_k = corr_with_target.abs().sort_values(ascending=False)[:k].index
selected_features = data[top_k]
```

```python
selected_corr_matrix = selected_features.corr()
selected_corr_matrix
```

```python
display(selected_features)
```

```python
selected_features.to_csv('../dataset-versions/murders-feature-selected.csv', index=False)
```
